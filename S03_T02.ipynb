Repeteix l'exercici 1 amb el dataset que disposem en el repositori de GitHub PRE-PROCESSING-DATA, movies.dat

```python
movies = pd.read_csv('movies.dat', sep='::', encoding = "ISO-8859-1")
movies.head()
```

```python
# Crear header con nombres de columnas sin borrar ninguna fila
primera_fila = [[i for i in movies.columns]]
movies.columns = ["ID","TITLE", "GENRES"]
movies = movies.append(pd.DataFrame( primera_fila,
               columns=movies.columns),
               ignore_index = True)


# Crear columna de a√±o
movies[['TITLE', 'YEAR']] = movies['TITLE'].str.split('(', 1, expand=True)


# Limpiar informaci√≥n
movies['TITLE'] = movies.TITLE.str.strip(' ')

movies['YEAR'] = movies.YEAR.str.strip(')')
movies['YEAR'] = movies.YEAR.str[-4:]
movies.loc[:,'YEAR'] = movies['YEAR'].astype(int)

movies['GENRES'] = [i for i in movies.GENRES.str.split('|')]

s = movies['GENRES'].explode()
print(s)
movies = movies.join(pd.crosstab(s.index, s))

movies.head()
```

## Exercici 3
En aquest exercici no us donarem gaires indicacions perqu√® volem que ens mostreu la vostra creativitat. Sorpr√®n-me amb gr√†fiques i interpretacions del dataset "movies.dat" del exercici anterior.

**Variable num√©rica y variable categ√≥rica:** Cantidad de pel√≠culas por g√©nero

```python
genres = [i for i in s]
genres = set(genres)
genres = list(genres)

genres_count = []
for i in genres:
    genres_count.append(movies[f'{i}'].sum())

genres_count_tuples = list(zip(genres_count,genres))

df_genres = pd.DataFrame(genres_count_tuples, columns=['Counts', 'Genre'])

sns.barplot(x="Counts", y="Genre", data=df_genres)
plt.title("Movies by genre")
plt.show()
```

**Variable categ√≥rica:** A√±o

```python
sns.set(style="whitegrid")
sns.boxplot(x = movies['YEAR'], color='violet', fliersize=2, orient='v', linewidth=1, width=0.3)
plt.title("Movies by year")
plt.show()
```

**Variables categ√≥ricas:** Comparaci√≥n g√©neros por a√±o, agrupados por afinidad

```python
movies_year = movies.groupby('YEAR').sum()

x = range(movies['YEAR'].min(),movies['YEAR'].max())
plt.stackplot(x, movies_year['Mystery'], movies_year['Crime'], movies_year['Thriller'], movies_year['Film-Noir'], movies_year['Horror'], labels=['Mystery', 'Crime', 'Thriller', 'Film-Noir', 'Horror'])
plt.legend(loc='upper left')
plt.show()

x = range(movies['YEAR'].min(),movies['YEAR'].max())
plt.stackplot(x, movies_year['Western'], movies_year['War'], movies_year['Adventure'], movies_year['Action'], labels=['Western', 'War', 'Adventure', 'Action'])
plt.legend(loc='upper left')
plt.show()

x = range(movies['YEAR'].min(),movies['YEAR'].max())
plt.stackplot(x, movies_year['Fantasy'], movies_year['Sci-Fi'], labels=['Fantasy', 'Sci-Fi'])
plt.legend(loc='upper left')
plt.show()

x = range(movies['YEAR'].min(),movies['YEAR'].max())
plt.stackplot(x, movies_year['Drama'], movies_year['Romance'], movies_year['Comedy'], labels=['Drama', 'Romance', 'Comedy'])
plt.legend(loc='upper left')
plt.show()

x = range(movies['YEAR'].min(),movies['YEAR'].max())
plt.stackplot(x, movies_year['Children\'s'], movies_year['Musical'], movies_year['Animation'], labels=['Children\'s', 'Musical', 'Animation'])
plt.legend(loc='upper left')
plt.show()

x = range(movies['YEAR'].min(),movies['YEAR'].max())
plt.stackplot(x, movies_year['Documentary'], labels=['Documentary'])
plt.legend(loc='upper left')
plt.show()
```

**Varible num√©rica y variable categ√≥rica:** Cantidad de pel√≠culas para ni√±os por g√©nero

```python
children_movies = movies[(movies['Children\'s'] == 1)]

children_genres = [i for i in s]
children_genres = set(children_genres)
children_genres = list(children_genres)

children_genres_count = []
for i in children_genres:
    children_genres_count.append(children_movies[f'{i}'].sum())

children_genres_count_tuples = list(zip(children_genres_count,children_genres))

df_children_genres = pd.DataFrame(children_genres_count_tuples, columns=['Counts', 'Genre'])
df_children_genres = df_children_genres.drop(11)

sns.barplot(x="Counts", y="Genre", data=df_children_genres)
plt.title("Children\'s movies by genre")
plt.show()
```
